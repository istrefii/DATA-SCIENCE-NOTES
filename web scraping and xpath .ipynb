```python
#import modules 
import requests 
import pandas as pd
import sqlite3
import lxml.html
from IPython.core.display import display, HTML
```

```python
# request get on the url
h1= requests.get('https://pokemondb.net/pokedex/all')
```

```python
# using the lxml.html function on the request get above
d1= lxml.html.fromstring(h1.content)
```

```python
#getting the table by the id
pokedex=d1.get_element_by_id('pokedex')
```

```python
pokedex
```

```python
# getting the children of the tablebody
tablebody= pokedex.getchildren()[1]
tablebody.getchildren()[0].text_content()
```

```python
# trying to find the first index of the table
tablebody.getchildren()[0].getchildren()[0].text_content()
```

```python
# getting the length of the table
len(tablebody.getchildren()[0:])
```

```python
#creating a dictionary with all the column names
data = {
    "#": [],
    "Name":     [],
    "Type":     [],
    "HP":       [],
    "Attack":   [],
    "Defense":  [],
    "Sp_Atk":   [],
    "Sp_Def":   [],
    "Speed":    [],
}
# creating a for loop to get each value in the table
for row in tablebody.getchildren()[0:]:
    data["#"].append(row.getchildren()[0].text_content().strip())
    data["Name"].append(row.getchildren()[1].text_content().strip())
    data["Type"].append(row.getchildren()[2].text_content().strip())
    data["HP"].append(row.getchildren()[3].text_content().strip())
    data["Attack"].append(row.getchildren()[4].text_content().strip())
    data["Defense"].append(row.getchildren()[5].text_content().strip())
    data["Sp_Atk"].append(row.getchildren()[6].text_content().strip())
    data["Sp_Def"].append(row.getchildren()[7].text_content().strip())
    data["Speed"].append(row.getchildren()[8].text_content().strip())
# creating the data frame
df= pd.DataFrame(data)
df
```

```python
# creating the db 
db = sqlite3.connect(":memory")
# creating the cursor to the db
cursor = db.cursor()
# executing the table, calling it PKS_TABLE with the columns 
cursor.execute("""
    CREATE TABLE PKS_TABLE(NUMBER, NAME, TYPE, HP, ATTACK, DEFENSE, SP_ATK, SP_DEF, SPEED)
""")
#using the for loop to insert the values
for row in df[df.Type.isin(["Fire", "Water"])].itertuples():
    insert_sql_syntax = """
        INSERT INTO PKS_TABLE(NUMBER, NAME, TYPE, HP, ATTACK, DEFENSE, SP_ATK, SP_DEF, SPEED)
        VALUES (?,?,?,?,?,?,?,?,?)
    """
    cursor.execute(insert_sql_syntax, row[1:])
    print (row[1:])
# db.close()
```

```python
# setting all the data where type=fire 
cursor.execute("""
    SELECT
    *
    FROM 
    PKS_TABLE
    WHERE
    TYPE='Fire'
""").fetchall()
```
